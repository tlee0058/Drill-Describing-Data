```python
import numpy as np
import random

binomialStat = np.random.binomial(50, 0.35, 100)
normalStat = np.random.normal(10, 1, 100)
poissonStat = np.random.poisson(1, 1000)
gammaStat = np.random.gamma(10, 1, 100)
gumStat = np.random.gumbel(10, 1, 100)
laStat = np.random.laplace(10,1,100)

def display(graph):
    %matplotlib inline
    from matplotlib import pyplot as plt
    plt.hist(graph)
    plt.axvline(graph.mean(), color='m', linestyle='solid', linewidth=2)
    plt.axvline(graph.mean() + graph.std(), color='b', linestyle='dashed', linewidth=2)
    plt.axvline(graph.mean() - graph.std(), color='b', linestyle='dashed', linewidth=2)
    
    plt.show()
    
display(binomialStat)
display(normalStat)
display(poissonStat)
display(gammaStat)
display(gumStat)
display(laStat)

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import random

rand1 = np.random.normal(5, 0.5, 100)
rand2 = np.random.normal(10, 1, 100)
rand3 = rand1 + rand2

plt.hist(rand3)
plt.axvline(rand3.mean(), color='m', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean() - rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```
